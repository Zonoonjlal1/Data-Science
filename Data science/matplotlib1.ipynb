```python
emplooy_name = ['Ahmed','Fadye','Sara','Dlal','Rvar']
Salary = [5000,3000,1950,4000,3680]
```

```python
plt.Figure(figsize=(15,5))
plt.plot(emplooy_name, Salary)
plt.title("emplooy")
plt.ylabel("emplooy_name")
plt.xlabel('Salary')
plt.show()
```

```python
plt.Figure(figsize=(15,5))
plt.bar(emplooy_name, Salary)
plt.title("emplooy")
plt.ylabel("emplooy_name")
plt.xlabel('Salary')
plt.show()
```
